```python
#GITHUB PROJECT 
#Ship Fuel Consumption & CO2 Emissions Analysis
#Analysis

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('ship_fuel_efficiency_cleaned.csv')
df
```

```python
#Calculate fuel consumption per mile
df['fuel_per_mile'] = df['fuel_consumption'] / df['distance']
```

```python
#Calculate the energy-consuming ship type 

ship_efficiency = df.groupby('ship_type')['fuel_per_mile'].mean().sort_values(ascending=False)
ship_efficiency
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(data=df, x='ship_type', y='fuel_per_mile', palette='Set2')

plt.title('Κατανάλωση Καυσίμου ανά Μίλι για κάθε Τύπο Πλοίου', fontsize=14)
plt.xlabel('Τύπος Πλοίου', fontsize=12)
plt.ylabel('Κατανάλωση ανά Μίλι', fontsize=12)
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x='fuel_consumption',
    y='CO2_emissions'
)

plt.title('Κατανάλωση καυσίμου έναντι εκπομπών CO2')
```

```python
sns.boxplot(
    data=df,
    x='weather_conditions',
    y='fuel_consumption'
)
plt.title('Κατανάλωση καυσίμου ανά καιρικές συνθήκες')
```

```python


# CO2 per mile by fuel type
df['CO2_per_mile'] = df['CO2_emissions'] / df['distance']
fuel_comparison = df.groupby('fuel_type')['CO2_per_mile'].mean()
fuel_comparison
```

```python
# CO2 Emissions per Month
plt.figure(figsize=(10,6))
monthly_co2 = df.groupby('month')['CO2_emissions'].sum().reindex([
    'January', 'February', 'March', 'April', 'May', 'June', 
    'July', 'August', 'September', 'October', 'November', 'December'])
monthly_co2.plot(kind='line', marker='o', color='green')
plt.title('Total Monthly Carbon Footprint (CO2)')
plt.savefig('slide6_monthly_co2.png')
plt.ylabel('CO2 Emissions')
```

```python
expensive_routes = df.groupby(['route_id', 'month', 'weather_conditions'])['fuel_per_mile'].mean().reset_index()

expensive_routes = expensive_routes.sort_values(by='fuel_per_mile', ascending=False)

expensive_routes
```

```python
#Engine Efficiency Analysis

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='engine_efficiency', y='fuel_consumption', hue='ship_type', alpha=0.7)
# Προσθήκη γραμμής τάσης (trendline) για να φανεί καθαρά η σχέση
sns.regplot(data=df, x='engine_efficiency', y='fuel_consumption', scatter=False, color='red')
plt.title('Engine Efficiency vs Fuel Consumption')
plt.xlabel('Engine Efficiency (%)')
plt.ylabel('Fuel Consumption')
plt.show()
```

```python
#SHIP TYPE FUEL

df['CO2_per_distance'] = df['CO2_emissions'] / df['distance']
```

```python
comparison_table = df.groupby('fuel_type')['CO2_per_distance'].mean().reset_index()
print("--- Πίνακας Σύγκρισης (Μέσος Όρος CO2/km) ---")
print(comparison_table)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='fuel_type', y='CO2_per_distance', ci=None, palette='muted')

plt.title('Σύγκριση Εκπομπών CO2 ανά Χιλιόμετρο: Diesel vs HFO')
plt.xlabel('Τύπος Καυσίμου')
plt.ylabel('Μέσο CO2 ανά Μονάδα Απόστασης')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

ax = sns.boxplot(
    x="weather_conditions",
    y="fuel_consumption",
    data=df,
    palette="coolwarm",
    hue="weather_conditions",
    legend=False,
)


plt.title(
    "Το Κόστος της Κακοκαιρίας: Κατανάλωση Καυσίμου ανά Καιρικές Συνθήκες",
    fontsize=14,
    fontweight="bold",
    pad=15,
)
plt.xlabel("Καιρικές Συνθήκες", fontsize=12, fontweight="bold")
plt.ylabel("Κατανάλωση Καυσίμου (Fuel Consumption)", fontsize=12, fontweight="bold")


plt.text(
    x=0.5,
    y=df["fuel_consumption"].max() * 0.85,
    fontsize=11,
    color="darkred",
    weight="bold",
    bbox=dict(facecolor="white", alpha=0.8, edgecolor="red", boxstyle="round,pad=0.5"),
)


plt.tight_layout()
plt.show()
```

```python
corr = df.select_dtypes(include='number').corr()

plt.figure(figsize=(10,8))

sns.heatmap(
    corr,
    annot=True,
    cmap='crest',
    fmt='.2f',
    linewidths=0.5,
    square=True
)

plt.title('Correlation Analysis', fontsize=16)

plt.tight_layout()
plt.show()
```

```python
avg_fuel_weather = df.groupby('weather_conditions')['fuel_consumption'].mean()

plt.figure(figsize=(8,5))
sns.barplot(
    x=avg_fuel_weather.index,
    y=avg_fuel_weather.values,
    palette='Blues_r'
)

plt.title('Average Fuel Consumption by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Average Fuel Consumption')
plt.show()
```
