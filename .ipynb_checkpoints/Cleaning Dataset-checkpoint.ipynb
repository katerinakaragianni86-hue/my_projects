```python
#GITHUB PROJECT 
#Ship Fuel Consumption & CO2 Emissions Analysis
#Data Cleaning

import pandas as pd
import numpy as np

raw_df = pd.read_csv('ship_fuel_efficiency.csv')
raw_df
```

```python
raw_df.shape
```

```python
raw_df.info()
```

```python
#Kaθαρισμός δεδομένων απο διπλοτυπές τιμές 

raw_df.duplicated().sum()
```

```python
#Eλεγχος τιμών να μην είναι αρνητικές ή μηδενικές

raw_df = raw_df [raw_df ['distance'] > 0]
raw_df = raw_df [raw_df ['fuel_consumption'] > 0]
raw_df  = raw_df [raw_df ['CO2_emissions'] > 0]
raw_df  = raw_df [(raw_df ['engine_efficiency'] >= 0) & (raw_df ['engine_efficiency'] <= 100)]
```

```python
raw_df.to_csv('ship_fuel_efficiency_cleaned.csv', index=False)
```
